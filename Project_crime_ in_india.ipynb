```python
df_KIDNAPPING = df.loc[df['CRIME HEAD']=='KIDNAPPING and ABDUCTION OF CHILDREN']
df_KIDNAPPING
```

```python
df_KIDNAPPING.drop(df_KIDNAPPING.index[[-1,-2,-10]],inplace=True)
```

```python
G=sbn.catplot(x='STATE/UT',y='TOTAL',kind='bar',data=df_KIDNAPPING,aspect=2)
G.set_xticklabels(rotation=90)
```

Uttar pradesh has the highest no. of kidnapping followed by Biahr and then Maharshtra
